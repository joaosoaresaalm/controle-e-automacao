### Seções em dropdown

Recolher grandes blocos de texto pode tornar sua marcação muito mais fácil de enxergar

<details>
<summary>"Clique aqui para expandir"</summary>
Texto escondido
</details>

```
<details>
<summary>"Clique aqui para expandir"</summary>
Texto escondido
</details>
```

Esconder grandes blocos de texto Markdown

<details>
<summary>Para certificar-se de que o markdown seja renderizado corretamente na seção recolhida ..summary>

 1. Coloque uma ** linha vazia ** após a tag `<summary>` .
 2. *Insira sua sintaxe de markdown*
 3. Coloque uma ** linha vazia ** antes da tag `</details>` 
 
</details>

```
<details>
<summary>To certifique-se de que o markdown seja renderizado corretamente na seção recolhida ...</summary>

 1. Coloque uma ** linha vazia ** após a tag `<summary>` .
 2. *Insira sua sintaxe de markdown*
 3. Coloque uma ** linha vazia ** antes da tag `</details>` 
 
</details>
```
