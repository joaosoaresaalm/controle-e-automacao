# 1. Introdução aos Sistemas de Controle 

## 1.1 - Introdução

Um sistema de controle pode ser definido como um conjunto de equipamentos e dispositivos que gerenciam o comportamento de máquinas ou outros sistemas físicos. Para isso, é necessário realizar a modelagem matemática da planta, seguida do projeto do controlador.

Nesta seção, iremos detalhar um pouco sobre a definição de um sistema de controle, bem como seu histórico e alguns exemplos.

## 1.2 - Desenvolvimento
    
Um sistema de controle é um conjunto de componentes e subsistemas interligados com um objetivo principal: manter uma ou muitas variáveis de um processo dentro de valores preestabelecidos, necessitando de pouca ou nenhuma supervisão humana.

Sistemas de controle automático estão presentes em diversos setores da sociedade moderna. Máquinas-ferramenta, robôs, telecomunicações, sistemas e equipamentos militares, equipamentos médicos e hospitalares, processos produtivos, veículos terrestres, marítimos, aéreos e espaciais e produtos de consumo contam cada vez mais com algum ou vários sistemas de controle. Isso tem sido possível em grande parte graças ao desenvolvimento contínuo da eletrônica e da computação.

Melhorar a eficiência energética de equipamentos e processos, torná-los mais produtivos, rápidos, precisos e seguros, aumentar a qualidade dos produtos, diminuir a emissão de poluentes e o consumo de insumos são algumas das vantagens competitivas que justificam a aplicação de sistemas de controle.

Especialmente no setor industrial, no qual as empresas precisam se manter competitivas para sobreviver em um mercado global, sistemas de controle são usados há bastante tempo nos processos de fabricação e tem se tornado cada vez mais sofisticados e imprescindíveis.

A base para análise de um sistema é formada pelos fundamentos fornecidos pela teoria dos sistemas lienares, que supõe uma relação de caisa e efeito para os componentes de um sistema. ### 1.2.1 - Componentes de um sistema de controle

Podemos definir de uma maneira geral os principais componentes de um sistema de controle:

**Ponto de ajuste ou setpoint:** é o valor usado como entrada do sistema. É o valor que se pretende obter na saída do sistema e é usualmente abreviado por SP.

**Planta ou processo:** pode ser um circuito eletrônico, um equipamento, uma máquina-ferramenta, um veículo, um processo industrial, etc. É a parte do sistema cuja variável de saída queremos controlar de tal forma que o valor dessa saída seja igual ao do setpoint.

**Variável controlada ou variável do processo:** é a variável de saída do sistema e a qual desejamos controlar.

**Atuador:** é o componente que recebe um valor de controle na sua entrada, geralmente amplifica e converte esse valor, e aplica o valor convertido na entrada do processo.

**Variável manipulada:** é a variável de saída do atuador. Usualmente abreviada por MV (de manipulated variable).

**Sensor:** é o componente que faz a medição de uma variável em um ponto do sistema, geralmente a variável do processo, e transmite esse valor para outra parte do sistema.

**Erro:** é a diferença entre o setpoint e a variável do processo. Essa é a informação de entrada do controlador e é de fundamental importância para o sistema de controle.

**Controlador:** é o componente do sistema que, com base no valor do erro, calcula o valor que será aplicado na entrada do atuador.

**Distúrbio:** o distúrbio é uma variável indesejável, quase sempre presente e que altera a saída do processo. Um sistema de controle eficaz é aquele que consegue manter a variável do processo controlada apesar dos distúrbios.

### 1.2.3 - Histórico do controle automático

As primeiras aplicações do controle com retroação apareceram no desenvolvimento do mecanismo regulador de bóia na Grécia no período entre 300 a 1 a.C. O relógio d'agua de Ktesibios usava um regulador de bóia. Um lampião de óleo inventado por Philon aproximadamente em 250 a.C. usava um regulador de bóia para manter constante o nível do óleop combustível. Heron de Alexandria, que viveu no primeiro século d.C., publicou um livro intitulado Pneumatica, que ressaltava algumas formas de mecanismos de nível de água usando reguladores de bóia. Após isso, podemos conferir os marcos históricos de acordo com a tabela abaixo:

| Ano  | Descrição  |
|---|---|
| 1769  | Desenvolvimento da máquina a vapor e do regulador de esferas de James Watt.| 
|  1800 | O conceito de intercambialidade de partes manufaturadas de Eli Whitney foi demonstrado na fabricação de mosquetões.|
| 1868  | J.C. Maxwell formula um modelo matemático para controle regulador de uma máquina a vapor. |
|  1913 | Introdução da máquina de montagem mecanizada de Henry Ford para a produção automobilística.  |
| 1927  | H. W. Bode analisa amplificadores com retroação.  |

### 1.2.4 - Objetivos de um sistema de controle

Já definimos anteriormente, ainda que de forma breve, o objetivo de um sistema de controle. Agora essa ideia será detalhada um pouco mais dividindo em três os principais objetivos a serem perseguidos no projeto de um sistema de controle.

#### 1.2.4.1 - Estabilidade

Esse é o principal objetivo de um sistema de controle. Se esse objetivo não puder ser alcançado então o sistema precisa ser reprojetado. Um sistema estável é aquele em que a variável de processo tende a atingir o valor de setpoint e a permanecer nesse valor ao longo do tempo.

Um sistema instável é aquele em que a variável de processo tende a oscilar constantemente, inclusive podendo aumentar a amplitude da oscilação.

#### 1.2.4.2 - Resposta transitória adequada

Resposta transitória é a resposta característica de um sistema após receber uma perturbação e até atingir a estabilidade. A resposta transitória é um requisito importante no sentido de contribuir para eliminar o erro o mais rápido possível.

Um sistema estável mas com resposta transitória lenta pode não ser adequado a determinadas aplicações. Por outro lado, respostas muito rápidas podem prejudicar a estabilidade. A resposta transitória precisa então ser ajustada para a minimizar o tempo de correção do erro sem prejudicar a estabilidade.

#### 1.2.4.3 - Eliminação do erro em regime permanente (ou offset)

O erro em regime permanente é a diferença entre o setpoint e a PV depois da estabilização do sistema. Ou, em outras palavras, é o erro que persiste na saída do sistema após terminada a resposta transitória. O erro em regime permanente é também chamado de offset.

Mesmo que um sistema possua estabilidade e boa resposta transitória, manter um erro em regime permanente não só é indesejável como pode ser inadmissível em determinados processos.

Imaginemos, por exemplo, um sistema que precisa manter a proporção exata na dosagem de dois produtos químicos diferentes.

Se o controle de um dos produtos tiver um erro em regime permanente a proporção não será mantida e o processo poderá ficar comprometido, seja em termos de qualidade, de consumo ou mesmo em termos de segurança dependendo dos produtos envolvidos.

### 1.2.5 - O projeto de Engenharia e Sistemas de Controle

É notória a necessidade de se saber projetar para um Engenheiro, aqui discutiremos o que vem a ser um Projeto de Engenharia e especificamente um projeto de um sistema de controle.

#### 1.2.5.1 - Projeto de Engenharia

O projeto de engenharia é a tarefa principal do engenheiro. Trata-se de um procedimento complexo no qual a criatividade e a análise desempenham os papéis principais. Projetar é o processo de conceber ou inventar formas, partes e detalhes de um sistema para alcançar um fim específico.

A atividade de projeto pode ser vista como um planejamento para dar lugar a umnovo produto ou sistema. O projeto é uma ação inovadora onde o engenheiro usa de forma criativa conhecimentos e materiais para especificar a forma, a função e o conteúdo material de um sistema. As etapas de um projeto são:
    1 - Determinar uma necessidade oriunda dos valores de vários grupos, cobrindo o espectro que vai desde os responsáveis pela elavoração das políticas públicas até os consumidores;
    2 - Especificar em detalhes o que deveria ser uma solução para essa necessidade e incorporar estes valores.
    3 - Desenvolver e avaliar várias soluções alternativas para antigir estas especificações
    4 - Decidir qual delas será projetada em detalhes de fabrica.
    
#### 1.2.5.2 - Projeto de Sistemas de Controle

O projeto de sistemas de controle é um exemplo específico de um projeto de engenharia. Aqui, novamente, a finalidade do projeto de engenharia de controle é obter a configuração, as especificações e a identificação dos parâmetros-chave de um sistema proposto para atender a uma necessidade real.

O primeiro passo no procedimento de projeto consiste em estabelecer a finalidade do sistema. O segundo passo é identificar as variáveis que se deseja controlar. O terceiro passo é escrever as especificações em termos da exatidão que se deve alcançar. ## 1.3 - Exemplos de Projeto

### 1.3.1 - Controle de velocidade de caminhão de uma empresa de entregas

Suponhamos que empiricamente concluíram que para obter a maior eficiência de consumo de combustível, um caminhão deve trafegar o maior tempo possível na velocidade de 105 km/h.

Porém a velocidade do caminhão depende de uma combinação de fatores externos como o peso da carga transportada, a inclinação da rodovia em diferentes trechos (subida, descida e plano), chuva, ventos contrários, etc. O motorista então deve estar atento ao velocímetro para compensar esses fatores acelerando mais ou menos e assim e procurar manter o veículo na velocidade de 105 km/h.

Para analisar esse sistema vejamos o seguinte diagrama de blocos:

<p align="center" text-align="center">
  <img width="500" height="900" src="img/cap1/6.png">
</p>

O diagrama acima representa o sistema de controle de velocidade do caminhão. A função desse sistema é fazer com que a velocidade real, que é a saída do sistema, seja igual ou o mais próximo possível da velocidade desejada, que é a entrada do sistema. Para isso o motorista precisa comparar constantemente a velocidade desejada com a real, o que ele consegue fazer através da informação fornecida pelo velocímetro.

Observe que a velocidade real varia em virtude da aceleração imposta ao caminhão e também dos fatores externos que podem mudar ao longo da viagem.

Se o velocímetro indica velocidade menor do que a desejada o motorista sabe que precisa pisar mais no acelerador. Se for o contrário o motorista sabe que precisa pisar menos. A atuação do motorista sobre o acelerador faz com que varie a velocidade do caminhão.

Ao atuar no acelerador o motorista verifica novamente o velocímetro para decidir se precisará pisar mais, menos (e quanto) ou manter a aceleração como está para controlar a velocidade real do caminhão.

### 1.3.2 - Sistema de Leitura do Acionador de Disco

Informações podem ser rápida e eficientemente armazenadas em discos magnéticos. Os acionadores de disco são usados em computadores portáteis (notebooks) e em computadores de todos os tamnhaos. As vendas mundiais de acionadores de disco foi estimada em 100 milhões de unidades em 1996. O objetivo do dispositivo leitor do acionador de disco é posicionar a cabeça de leitura a fim de que sejam lidos os dados armazenados em uma das trilhas do disco. A variável para controlar precisamente é a posição da cabeça leitora.

A especificação inicial para a precisão do posicionamento é de 1um. Além disso, planeja-se mover a cabeça da trila a para a trilha b dentro de 50ms, se possível. Assim, fica estabelecida uma configuração inicial de sistema. Este sistema a malha fechada proposto usa um motor para acionar o braço até o local desejado no disco.
